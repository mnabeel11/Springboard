```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import math as m
import warnings 
warnings.filterwarnings(action="ignore")
dfs = pd.read_csv("D:\Download_Ddrive\default.csv",skiprows=1)
dfs=dfs.rename(index=str, columns={"default payment next month": "default"})
dfs.head()
```

```python
# Hypothesis Test: Difference Between Proportions
# Two sided z-proportion test to see if the proportion of male and female defaulters is equal.
# null hypothesis is that proportion of male and female defaulters is equal.

dfs_sex_tot=dfs.SEX.count()
print("Total no. of males and females: " + str(dfs_sex_tot))

dfs_sex_def_tot=dfs[dfs.default==1].SEX.count()
print("Total no. of defaulters: " + str(dfs_sex_def_tot))

dfs_female=dfs[dfs.SEX==2][dfs.default==1].SEX.count()
print("Total no. of female defaulters: " + str(dfs_female))

dfs_male=dfs[dfs.SEX==1][dfs.default==1].SEX.count()
print("Total no. of males defaulters: " + str(dfs_male))

prop_male_def=dfs_male/dfs[dfs.SEX==1].SEX.count()
print("Proportion of male defaulters: " + str(prop_male_def))


prop_female_def=dfs_female/dfs[dfs.SEX==2].SEX.count()
print("Proportion of female defaulters: " + str(prop_female_def))

prop_total=dfs_sex_def_tot/dfs_sex_tot

upper=prop_male_def-prop_female_def
lower=m.sqrt(prop_total*(1-prop_total)*((1/dfs[dfs.SEX==2].SEX.count())+(1/dfs[dfs.SEX==1].SEX.count())))
z_value=upper/lower
print("z-value is : "+str(z_value))
p_value =stats.norm.sf(abs(z_value))*2 #twosided
```

* We chose significance level of 0.05. 

* Since the p_value is less than the significance value we reject the null hypothesis that the proportion of male and female 
  defaulters is equal 


```python
# Is credit limit a significant predictor of default 
import statsmodels.formula.api as sm
result = sm.ols(formula='default ~ LIMIT_BAL', data=dfs).fit()
print(result.summary())
```

* Since the p_value is less than 0.05 we can safely say that credit limit is a good predictor of default. 


```python
# Is BILL_AMOUNT a significant predictor of default ?
import statsmodels.formula.api as sm
result = sm.ols(formula='default ~ BILL_AMT1+BILL_AMT2+BILL_AMT3+BILL_AMT4+BILL_AMT5+BILL_AMT6', data=dfs).fit()
print(result.summary())
```

* Since the  p value for BILL_AMT1 is the only one that is less than 0.05 significance level we reject our hypothesis that 
  BILL amount overall has anything to do with the default of credit card clients. 
  

```python
# Is AGE a significant predictor of default ?

import statsmodels.formula.api as sm
result = sm.ols(formula='default ~ AGE', data=dfs).fit()
print(result.summary())
```

* Since our p-value is less than 0.05 for age we can continue with our hypothesis that age does play a role in default. 


```python
# Which Age limits have the highest significance of default

dfs_age_21=dfs["AGE"][dfs.AGE <= 21]
dfs_age_40=dfs["AGE"][dfs.AGE >= 21][dfs.AGE<=40]
dfs_age_60=dfs["AGE"][dfs.AGE >=40][dfs.AGE <=60]
dfs_age_80=dfs["AGE"][dfs.AGE >=60][dfs.AGE <=80]
dfs_default_21=dfs['default'][dfs.AGE <=21]
dfs_default_40=dfs['default'][dfs.AGE >=21][dfs.AGE<=40]
dfs_default_60=dfs['default'][dfs.AGE >=40][dfs.AGE<=60]
dfs_default_80=dfs['default'][dfs.AGE >=60][dfs.AGE<=80]

import statsmodels.formula.api as sm
result21 = sm.ols(formula='dfs_default_21 ~ dfs_age_21 ', data=dfs.AGE).fit()
print(result21.summary())


import statsmodels.formula.api as sm
result40 = sm.ols(formula='dfs_default_40 ~ dfs_age_40 ', data=dfs.AGE).fit()
print(result40.summary())

import statsmodels.formula.api as sm
result60 = sm.ols(formula='dfs_default_60 ~ dfs_age_60 ', data=dfs.AGE).fit()
print(result60.summary())

import statsmodels.formula.api as sm
result80 = sm.ols(formula='dfs_default_80 ~ dfs_age_80 ', data=dfs.AGE).fit()
print(result80.summary())
```

RESULTS: 

1. Ages between 0 and 21 show age as a strong predictor of default. 
2. Ages between 21 and 40 show age as a strong predictor of default. 
3. Ages between 40 and 60  show age as a strong predictor of default.
4. Ages between 60 and 80 show age as not a good predictor of default since our p-value is greater than significance value.
